```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Day_12_banking_data.csv')

# Bar plot for total sum of Transaction_Amount per Account_Type
account_type_sum = df.groupby('Account_Type')['Transaction_Amount'].sum()
account_type_sum.plot(kind='bar', title='Total Transaction Amount per Account Type')
plt.xlabel('Account Type')
plt.ylabel('Total Transaction Amount')
plt.show()

# Pie chart for percentage of transactions per Branch
branch_count = df['Branch'].value_counts()
branch_count.plot(kind='pie', autopct='%1.1f%%', title='Percentage of Transactions per Branch', figsize=(7, 7))
plt.ylabel('')
plt.show()
```
